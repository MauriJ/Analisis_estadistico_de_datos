```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
rng = np.random.default_rng()
from scipy.special import factorial
from scipy.stats import poisson
```

```python
ext_li = 12.7
ext_ls = 31.8
anillo_doble = 170
num_tiradas = 100
num_corridas = 10000
```

```python
def condicion(x):
    return x > ext_li and x < ext_ls

t = []

for i in range(num_corridas):
    r = tiradas = rng.integers(0, high=anillo_doble, size=num_tiradas, endpoint=True)
    count = sum(1 for x in r if condicion(x))
    t.append(count)
  
hist, bins = np.histogram(t, bins = np.linspace(0,30,16)) 

n = num_tiradas
p = (ext_ls - ext_li)/anillo_doble

mu = n*p
kmin = poisson.ppf(0.01, mu)
kmax = poisson.ppf(0.99, mu)
k = np.arange(kmin, kmax)

plt.hist(t, bins, alpha=0.75, edgecolor='black', density = True)
plt.plot(k, poisson.pmf(k, mu), 'o')
plt.show()
```
