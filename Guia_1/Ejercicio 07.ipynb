```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
rng = np.random.default_rng()
from collections import Counter
from scipy.stats import binom
```

```python
frecuencia = []

for i in range(1, 10000):
    tiradas = rng.integers(1, high=6, size=5, endpoint=True)
    cuentas = Counter(tiradas)
    num_tres = cuentas[3]
    frecuencia.append(num_tres)  

bin = np.arange(0,6,1)
hist, bins = np.histogram(frecuencia, bins = bin)

n = 5
p = 1/6

kmin = binom.ppf(0.01, n, p)
kmax = binom.ppf(0.99, n, p)
k = np.arange(kmin, kmax+2)
rv = binom(n, p)

plt.figure(figsize = (7.5,5))
plt.hist(frecuencia, bins, alpha=0.75, edgecolor='black', density = True)
plt.plot(k+0.5, binom.pmf(k, n, p), 'o', markersize = 10)
plt.xlabel('x', fontsize = 14)
plt.ylabel('f(x)', fontsize = 14)
plt.show()
```

```python
plt.figure(figsize=(7.5,5))
plt.plot(frecuencia)
plt.axhline(5/6, linestyle = '--', color = 'k')
plt.xlabel('Numero de simulaci√≥n', fontsize = 14)
plt.ylabel('Frecuencia de ocurrencia #3', fontsize = 14)
plt.show()
```
