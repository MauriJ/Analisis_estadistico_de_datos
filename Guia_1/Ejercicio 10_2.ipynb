```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
from scipy.stats import poisson
from scipy.stats import binom
rng = np.random.default_rng()
```

```python
mu = 10
p = 0.7

Z = []

for i in range(10000):
    Y = rng.poisson(mu)
    X = rng.binomial(1, p, Y)
    Z.append(X.sum())

mu2 = mu*p
```

```python
## Construyo los histogramas        

bin = np.arange(0,25,1)
hist_Z, bins_Z = np.histogram(Z, bins = bin)  

plt.hist(Z, bins_Z, alpha=0.5, density = True, color = 'C1', label = 'MC Poisson')
plt.plot(bin+0.5, poisson.pmf(bin, mu2), 'o', color = 'C3', markersize = 10, label = 'Simulacion Poisson')
plt.xlim(0, 20)
plt.xlabel('x', fontsize = 14)
plt.ylabel('P (x)', fontsize = 14)
plt.legend(loc = 'best')
plt.show()
```
