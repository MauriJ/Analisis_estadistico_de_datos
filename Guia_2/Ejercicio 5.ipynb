```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
from scipy.stats import norm
```

```python
mu = 1.7
num_variables = 10
num_sim = 1000

list_p = []

for i in range(num_sim):
    varx = poisson.rvs(mu, size=num_variables)
    vary = np.sum(varx)
    list_p.append(vary)

loc = np.mean(list_p)
scale = np.std(list_p)
```

```python
bin = np.linspace(0,40,20)
hist, bins = np.histogram(list_p, bins = bin)

x = np.linspace(0, 40, 256)
gauss = norm.pdf(x, loc, scale)

plt.hist(list_p, bins, alpha=0.5, density = True, label = 'Poisson')
plt.plot(x, gauss, 'r-', lw=5, label = 'Gaussiana')
plt.xlabel('X', fontsize = 18)
plt.ylabel('Densidad', fontsize = 18)
plt.legend(loc = 'best', fontsize = 14)
plt.show()
```
