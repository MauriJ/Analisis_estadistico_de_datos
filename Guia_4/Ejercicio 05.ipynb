```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 12
k = 8

x = np.arange(0, 1, 0.005)
L = 16.5*np.power(x,8)*np.power(1-x,4)

plt.plot(x, L)
plt.axvline(k/n, color="black", linestyle="--")
plt.xlabel('p', fontsize = 14)
plt.ylabel('L(p)', fontsize = 14)
plt.show()
```

```python
n = 12
k = 8

x = np.arange(0.005, 1, 0.005)
lam = np.power(x,k)*np.power(1-x,n-k)/(k/n)**k/(1-k/n)**(n-k)
J = -2*np.log(lam)

plt.plot(x, J)
plt.axvline(k/n, color="black", linestyle="--")
plt.xlabel('p', fontsize = 14)
plt.ylabel('J(p)', fontsize = 14)
plt.show()
```
