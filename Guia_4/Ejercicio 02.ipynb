```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
X = []

for i in range(1000):
    x_i = norm.rvs(loc=0, scale=1, size=3)
    X.append(np.mean(x_i))
```

```python
bin = np.arange(-3.5,3.5,0.1)
hist, bins = np.histogram(X, bins = bin)

plt.hist(X, bins, alpha=0.5, density = True, color = 'C0')
plt.plot(bin - 0.05, norm.pdf(bin - 0.05, 0, 1/np.sqrt(3)), 'o', color = 'C1', markersize = 10)
plt.xlim(-3, 3)
plt.xlabel('x', fontsize = 14)
plt.ylabel('P (x)', fontsize = 14)
# plt.legend(loc = 'best')
plt.show()
```
