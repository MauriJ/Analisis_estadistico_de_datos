```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [3.38, 5.06, 7.67]
s = 1.4
n = np.shape(X)[0]

mu = np.arange(np.mean(X)-10, np.mean(X)+10, 0.1)

suma = []

for i in np.arange(np.shape(X)[0]):
    suma.append((X[i] - mu)**2)

sumatoria = np.sum(suma, axis=0)

lnL = (-n/2)*np.log(2*np.pi*s**2) - (1/2/s**2)*sumatoria
```

```python
plt.plot(mu, lnL)
plt.axvline(np.mean(X), color="black", linestyle="--")
plt.xlabel('$\mu$', fontsize = 14)
plt.ylabel('lnL($\mu$)', fontsize = 14)
plt.show()
```

```python
suma = []

for i in np.arange(np.shape(X)[0]):
    suma.append((X[i] - np.mean(X))**2)

sumatoria = np.sum(suma, axis=0)

lnLmax = (-n/2)*np.log(2*np.pi*s**2) - (1/2/s**2)*sumatoria
lam = np.exp(lnL)/np.exp(lnLmax)
J = -2*np.log(lam)

J1 = ((np.mean(X) - mu)/(s/np.sqrt(n)))**2

plt.plot(mu, J, lw = 3, label = 'J por cociente L/L$_{max}$')
plt.plot(mu, J1, '--', lw = 3, label = 'J por ecuacion')
plt.axvline(np.mean(X), color="black", linestyle="--")
plt.xlabel('$\mu$', fontsize = 14)
plt.ylabel('J($\mu$)', fontsize = 14)
plt.legend(loc = 'best')
plt.show()
```
